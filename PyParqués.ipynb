```python
'''El tablero tiene 68 casillas en las que todas las fichas pueden avanzar numeradas desde 1 al 68 empezando
 por uno de los colores/equipos que escoja. Estas 68 casillas son todas las casillas externas de un tablero que se
 muestra, formalmente serían todas las casillas que colindan con uno de los cuatro cuadrados que están en las
 esquinas, excepto por la casilla de seguro que se encuentra en medio. 8 casillas para cada equipo/jugador que
 son exclusivas para cada uno de estos que se encuentra en las casillas Internas inmediatamente anterior a la
 casilla de salida de cada uno de los equipos. Algunas Reglas: Las fichas salen a su respectiva salida que está
 exactamente a 5 casillas desde el seguro inmediatamente anterior, Cada ficha sale con un cinco en los dados,
 ya sea repartido entre los dos dados o con el dígito completo en uno de los dos dados. Solo se permiten dos fichas
 máximo por cada casilla. En caso de que ya haya dos fichas en la salida de un respectivo equipo, las fichas solo
 podrán hacer movimientos. No obstante, las fichas también pueden moverse y desocupar el espacio para que salga
 otra ficha. Es obligatorio sacar una ficha de la cárcel cuando un 5 lo permita. Si una ficha llega a su llegada,
 entonces se pueden efectuar 10 movimientos con cualquier ficha de dicho equipo, siempre y cuando sea posible,
 considerando las reglas anteriores. Si un equipo saca tres pares consecutivos, la última ficha que haya movido
 deberá regresar a la cárcel. Un equipo ganará si todas sus fichas están en la llegada.'''

import random
dice_CaraDado = {


   6: [
        "┌─────────┐",
        "│ ●  ●  ● │",
        "│         │",
        "│ ●  ●  ● │",
        "└─────────┘"  ],
   5: [
        "┌─────────┐",
        "│ ●     ● │",
        "│    ●    │",
        "│ ●     ● │",
        "└─────────┘"
    ],
    4: [
        "┌─────────┐",
        "│ ●     ● │",
        "│         │",
        "│ ●     ● │",
        "└─────────┘"
    ],

    3: [
        "┌─────────┐",
        "│       ● │",
        "│    ●    │",
        "│ ●       │",
        "└─────────┘"
    ],
    2: [
        "┌─────────┐",
        "│       ● │",
        "│         │",
        "│ ●       │",
        "└─────────┘"
    ],

    1: [
        "┌─────────┐",
        "│         │",
        "│    ●    │",
        "│         │",
        "└─────────┘"
    ],


}

class Ficha:
    def __init__(self, color, id_ficha, jugador):
        self.color = color
        self.id_ficha = id_ficha
        self.posicion = 'carcel'  # Puede ser 'carcel', número de casilla o 'llegada'
        self.jugador = jugador

    def salir_de_carcel(self, tablero):
        if self.posicion == 'carcel':
            casilla_salida = self.jugador.casilla_salida
            if tablero.mover_ficha(self, casilla_salida):
                self.posicion = casilla_salida
                print(f"Ficha {self.id_ficha} de {self.color} ha salido de la cárcel y está en la casilla {casilla_salida}.")
                return True
        return False

    def mover(self, pasos, tablero):
        if self.posicion == 'carcel':
            print(f"Ficha {self.id_ficha} de {self.color} No puede mover ficha porque está en la cárcel.")
            return False
        nueva_posicion = self.posicion + pasos
        if nueva_posicion > 68:
            print(f"Ficha {self.id_ficha} de {self.color} No puede mover ficha porque excede la meta.")
            return False
        if nueva_posicion == 68:
            print(f"Ficha {self.id_ficha} de {self.color} ¡Ya puede sacar su ficha! Se otorgan 10 movimientos extra.")
            self.jugador.movimientos_extra += 10
        if tablero.mover_ficha(self, nueva_posicion):
            self.posicion = nueva_posicion
            print(f"Ficha {self.id_ficha} de {self.color} movida a casilla {nueva_posicion}")
            return True
        return False

class Jugador:
    def __init__(self, color):
        self.color = color
        self.fichas = [Ficha(color, i, self) for i in range(4)]
        self.movimientos_extra = 0
        self.pares_consecutivos = 0
        self.casilla_salida = self.definir_salida()

    def definir_salida(self):
        salidas = {'rojas': 5, 'azules': 22, 'verdes': 39, 'blancas': 56}
        return salidas[self.color]

    def lanzar_dados(self, modo_desarrollador=False):
        if modo_desarrollador:
            while True:
                dado1 = int(input(f"Ingresa el valor del primer dado (1-6): "))
                dado2 = int(input(f"Ingresa el valor del segundo dado (1-6): "))
                if 1 <= dado1 <= 6 and 1 <= dado2 <= 6:
                    break
                print("Valores inválidos, ambos deben estar entre 1 y 6.")

            # Mostrar dados elegidos
            print("\nDados seleccionados:")
            for i in range(5):
                print(dice_CaraDado[dado1][i] + "    " + dice_CaraDado[dado2][i])
            return dado1, dado2

        else:  # Modo real
            dado1 = random.randint(1, 6)
            dado2 = random.randint(1, 6)
            # Mostrar resultado de los dados
            print(f"\nLa máquina ha lanzado los dados:")
            for i in range(5):
                print(dice_CaraDado[dado1][i] + "    " + dice_CaraDado[dado2][i])
            print(f"Valores: Dado 1 = {dado1}, Dado 2 = {dado2}.\n")
            return dado1, dado2

            #if dado1 == dado2:
        #    self.pares_consecutivos += 1
        #    if self.pares_consecutivos >= 3:
    def intentar_sacar_ficha(self, tablero, dado1, dado2):
        if 5 in [dado1, dado2] or dado1 + dado2 == 5:
           for ficha in self.fichas:
              if ficha.posicion == 'carcel':
                  return ficha.salir_de_carcel(tablero)
        return False

class Tablero:
    def __init__(self):
        self.casillas = {i: [] for i in range(1, 69)}
        self.seguro = [5, 12, 17, 22, 29, 34, 39, 46, 51, 56, 63, 68]

    def mover_ficha(self, ficha, nueva_posicion):
        if nueva_posicion > 68:
            return False
        if ficha.posicion != 'carcel':
            self.casillas[ficha.posicion].remove(ficha)
        self.casillas[nueva_posicion].append(ficha)
        return True

class Juguemos:
    def __init__(self, modo_desarrollador=False):
        self.tablero = Tablero()
        self.jugadores = []
        self.turno = 0
        self.modo_desarrollador = modo_desarrollador

    def agregar_jugador(self, color):
        self.jugadores.append(Jugador(color))

    def jugar_turno(self):
        print(f"Turno de las fichas {self.jugadores[self.turno].color}")
        jugador_actual = self.jugadores[self.turno]
        dado1, dado2 = jugador_actual.lanzar_dados(self.modo_desarrollador)
        print(f"Alternativas con resultados en: {dado1}, {dado2}")
        if not jugador_actual.intentar_sacar_ficha(self.tablero, dado1, dado2):
            fichas= [ficha for ficha in jugador_actual.fichas if isinstance(ficha.posicion, int)]
            if fichas:
                print("Fichas disponibles para mover:")
                for ficha in fichas:
                    print(f"Ficha {ficha.id_ficha} en casilla {ficha.posicion}")
                id_ficha = int(input("Seleccione el ID de la ficha a mover: "))
                ficha_a_mover = next((ficha for ficha in fichas if ficha.id_ficha == id_ficha), None)
                if ficha_a_mover:
                    pasos = int(input(f"Seleccione cuántos pasos mover la ficha {ficha_a_mover.id_ficha} ({dado1} o {dado2}): "))
                    while pasos not in [dado1, dado2]:
                        pasos = int(input("Número inválido. Debe ser igual a uno de los valores de los dados: "))
                    ficha_a_mover.mover(pasos, self.tablero)
        for ficha in jugador_actual.fichas:
            print(f"Ficha {ficha.id_ficha} está en casilla {ficha.posicion}")
        self.turno = (self.turno + 1) % len(self.jugadores)

def main():
    print("¡¡¡ BIENVENIDO(S) A SU JUEGO DE PARQUÉS VIRTUAL !!!\n")
    print("Modalidad para jugar\n 1: Real(azar)\n 2: programador(determinada por usuario)\n ")
    modo = input("Seleccione el modo de juego (1 o 2): ")
    modo_programador = modo == '2'

    if modo_programador:
        print("El Modo que seleccionó fue el de 'Programador'.")
    elif modo == '1':
        print("Le confirmamos que el modo que seleccionó fue el de 'real'.")
    else:
        print("\nPuede escoger solamente entre 1 o 2\n")
        main()
        return


    juego = Juguemos(modo_programador)
    colores_disponibles = ['rojas', 'azules', 'verdes', 'blancas']

    num_jugadores = int(input("\nElige el número de jugadores posibles desde 2 a 4 personas:"))
    for i in range(num_jugadores):
        color = input(f"-Escoge fichas del jugador{i+1} {colores_disponibles}: ")
        while color not in colores_disponibles:
            color = input("Color NO válido, seleccione otra vez: ")
        juego.agregar_jugador(color)
        colores_disponibles.remove(color)

    print("\nIniciemos la partida del parqués virtual:\n")
    while True:
        juego.jugar_turno()
        if input("\nPresione Enter si quiere seguir jugando, de lo contrario 'n' para detenerse,\ny luego enter para detener la partida: ") == 'n':
            break

if __name__ == "__main__":
    main()


```
